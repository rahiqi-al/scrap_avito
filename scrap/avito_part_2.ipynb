# all columns show positive skew which means they are asymmetric so we need to use median imputation 

```python
data_nemeric = ['prix','N_chambre','N_salles_bain','surface', 'etage']
df_median_imputation=df.copy()
for d_n in data_nemeric:
    percentage_before = (df_median_imputation[d_n].isna().sum() / len(df_median_imputation[d_n])) * 100
    df_median_imputation[d_n].fillna(df_median_imputation[d_n].median(), inplace = True)
    percentage_after = (df_median_imputation[d_n].isna().sum() / len(df_median_imputation[d_n])) * 100
    print(f'{d_n}:percentage of nan value {percentage_before} and after imputation{percentage_after}')
    print('---------------')
   
```

# outlier detection and treatment

```python
data_nemeric = ['prix','N_chambre','N_salles_bain','surface', 'etage']
df_outliers=df[[]]
for d_n in data_nemeric:
    Q1 = df_median_imputation[d_n].quantile(0.25)
    Q3 = df_median_imputation[d_n].quantile(0.75)
    IQR = Q3 - Q1
    Limite_inf= Q1 - 1.5 * IQR
    limite_sup= Q3 + 1.5  *IQR
    outliers_for_each_column = df_median_imputation[(df_median_imputation[d_n] < Limite_inf) | (df_median_imputation[d_n] > limite_sup)]
    df_outliers=pd.concat([df_outliers,outliers_for_each_column],ignore_index=True)
df_outliers.drop_duplicates(inplace=True)
    
    
```

```python
X = clean_normalized_df.drop('localisation', axis=1)
y = clean_normalized_df['localisation'].astype(str)

#One-hot encode categorical columns in X
X = pd.get_dummies(X, drop_first=True)

#Convert all boolean columns to integers (True -> 1, False -> 0)
X = X.apply(lambda col: col.astype(int) if col.dtype == 'bool' else col)

#Convert any non-numeric columns to numeric
X = X.apply(pd.to_numeric, errors='coerce')
X.fillna(X.mean(), inplace=True)

# Filter out classes with only one sample in the target column to avoid SMOTE errors
class_counts = y.value_counts()
classes_to_keep = class_counts[class_counts > 1].index  # Classes with more than one sample
X_filtered = X[y.isin(classes_to_keep)]
y_filtered = y[y.isin(classes_to_keep)]

#Apply SMOTE on the filtered data to balance classes in the target column 'localisation'
smote = SMOTE(random_state=42, k_neighbors=1)
X_res, y_res = smote.fit_resample(X_filtered, y_filtered)


```

```python
# Convert the  X_res to a DataFrame
X_res_df = pd.DataFrame(X_res, columns=X.columns)
X_res_df
```

```python
#Add the y_res as a new column 
X_res_df['localisation'] = y_res
```
